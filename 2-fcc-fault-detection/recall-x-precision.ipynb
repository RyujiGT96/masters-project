```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,10))
plt.scatter(0.997, 0.813, marker='o', s=200, color='green', label='Bayes-LSTM')
plt.scatter(0.715, 0.477, marker='X', s=200, color='black', label='LSTM-Gaussian Tail')
plt.scatter(0.483, 0.503, marker='P', s=200, color='red', label='LSTM-Dynamic Error Threshold')
plt.xlabel('Sensibilidade (TPR)', fontsize=12)
plt.ylabel('Precisão', fontsize=12)
plt.xlim((0,1.05))
plt.ylim((0,1.05))
plt.legend(loc=(0.025, 0.8), labelspacing=1.5)
plt.grid()
plt.show()
```
